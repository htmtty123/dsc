```python


import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
```

```python
main()
```

/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
def question10(conn):
    """
    Query a list of facilities with a total revenue less than 1000.
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT temp.name, sum(temp.rev) as tot_rev
        FROM
        (
        SELECT a.facid,a.name,a.membercost,a.guestcost,b.starttime,b.bookid,b.memid,b.slots,
        (CASE WHEN b.memid <> 0 then a.membercost * b.slots
        WHEN b.memid = 0 THEN a.guestcost * b.slots END) AS rev
        FROM facilities a
        INNER JOIN bookings b ON a.facid=b.facid
        ) temp
        GROUP BY temp.name
        HAVING sum(temp.rev) < 1000
        """
    
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main10():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        question10(conn)
    
```

```python
main10()
```

/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */

```python
def question11(conn):
    """
    Query a report of members and who recommended them in alphabetic surname,firstname order
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query11 = """
        SELECT (a.surname || ' ' || a.firstname) AS membername,
        (SELECT (b.surname || ' ' || b.firstname) FROM members b WHERE a.recommendedby = b.memid) AS recommended_by
        FROM members a
        ORDER BY membername
        """
    
    cur.execute(query11)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main11():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        question11(conn)
```

```python
main11()
```

 Q12: Find the facilities with their usage by member, but not guests

```python
def question12(conn):
    """
    Query facilities with their usage by member, but not guests
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query12 = """
        SELECT  DISTINCT b.facid, f.name, b.memid, (m.firstname ||' ' || m.surname) AS member_name
        FROM `Bookings` b
        INNER JOIN `Members` m
        ON b.memid = m.memid
        INNER JOIN `Facilities` f
        ON b.facid = f.facid
        WHERE b.memid != 0
        ORDER BY b.facid,b.memid
        """
    
    cur.execute(query12)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main12():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        question12(conn)
```

```python
main12()
```

```python
Q13: Find the facilities usage by month, but not guests 
```

```python
def question13(conn):
    """
    Query facilities with their usage by month, but not guests
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query13 = """
        SELECT  (SELECT b.name FROM facilities b WHERE b.facid = a.facid) AS fname, strftime('%m',a.starttime) AS MONTH,  (sum(slots)/2)  AS  hours_usage 
        FROM bookings a
        WHERE a.memid !=0
        GROUP BY fname,month
        ORDER BY fname,month
        """
    
    cur.execute(query13)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main13():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        question13(conn)
```

```python
main13()
```
